## Research Question
The decision this work aims to improve is deciding which content pages should be reviewed or refreshed first. The primary users are SEO specialists and content managers, who use the recommendations to prioritize their work. A wrong recommendation could waste editor time by refreshing healthy pages or allow declining pages to remain unchanged, potentially reducing search visibility and traffic. Machine learning is useful because many observable signals, such as content age, impressions, click-through rate, and average position, interact in ways that are difficult to capture with a single hand-written rule.
