## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

My chosen lane is **Content Refresh & SEO Optimization**.

This is a **classification** problem because the goal is to predict whether a webpage is likely to experience declining search performance.

The model classifies each page into categories such as:

- Down (needs attention)
- Up (performing well)

This helps prioritize which pages should be refreshed first.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

The ideal goal is to identify pages that should be refreshed.

Since the dataset does not contain a column called "Needs Refresh", I use **performance_label** as a proxy target.

The labels are based on observed page performance:

- Down = declining performance
- Up = stable or improving performance

This proxy allows the model to learn from historical outcomes instead of manually defined rules.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

My primary success metric is **Precision@K**.

The content team can only refresh a limited number of pages, so it is important that the highest-ranked pages predicted by the model truly need attention.

A good model should achieve high Precision@K while maintaining useful precision and recall.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

One row represents one webpage.

Each row contains measurements describing that webpage, including metrics such as impressions, clicks, CTR, average position, search volume, page age, and word count.

The model makes one prediction for each webpage.

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

A simple rule such as "refresh every page with CTR below 5%" ignores many other factors.

Search performance depends on several features working together, including impressions, ranking position, search volume, page age, clicks, and word count.

Machine learning can learn patterns from these combined features and make better predictions than a single if-statement. This provides better decision support for prioritizing content updates.
